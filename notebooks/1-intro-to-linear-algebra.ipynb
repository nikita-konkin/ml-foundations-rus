### Что такое линейная алгебра

При формализации интуитивно понятных концепций принято подбирать набор объектов (символов) и давать набор правил по обращению с этими объектами. Такая наука называется алгеброй. Линейная алгебра — это наука о векторах и определенных правилах операций над ними. Векторы, которые многие помнят из школьного курса, называются «геометрическими» и обычно обозначаются стрелочкой над буквой, например, $\vec{x}$ и $\vec{y}$.

В принципе, векторы — это объекты особого рода, которые можно складывать друг с другом и умножать на скаляры, чтобы получить новый объект того же рода. С точки зрения абстрактной математики, любой объект, обладающий двумя этими свойствами, может считаться вектором. Вот несколько примеров таких векторных объектов:

1. Геометрические векторы. Векторы такого рода изучаются в старших классах, в курсах математики и физики. Геометрические векторы — см. рис. 1 (а) — это направленные отрезки, которые можно чертить (как минимум в двух измерениях). Два геометрических вектора  $\vec{x}$ и $\vec{y}$, можно сложить, так что их суммой $\vec{z} = \vec{x} + \vec{y}$ будет третий геометрический вектор. Более того, умножение на скаляр, $\lambda \vec{x}$  также даст геометрический вектор. Фактически это исходный вектор, умноженный на $\lambda$.
2. Полиномы — это тоже векторы; см. рис. 1 (b): два полинома можно сложить друг с другом, получив в результате третий полином; также их можно умножать на скаляр $\lambda$, и в результате тоже получится полином. Следовательно, полиномы — это образцы векторов (пусть и довольно необычные). Обратите внимание на то, что полиномы очень отличаются от геометрических векторов. Тогда как геометрический вектор — это конкретный «рисунок», полином — это абстрактная концепция. Однако и те, и другие являются векторами в вышеизложенном смысле.
3. Элементы $\mathbb R ^ n$ (кортежи из n вещественных чисел) — это векторы. Покомпонентное сложение двух векторов $a, b \in  \mathbb R ^ n$ дает еще один вектор: $a + b = c \in  \mathbb R ^ n$. Более того, при умножении $a \in \mathbb R ^ n$ на $\lambda \in \mathbb R$ получается умноженный вектор $\lambda a \in \mathbb R ^ n$. Рассматривать векторы как элементы $\mathbb R ^ n$ удобно еще и потому, что в таком представлении они условно соответствуют массивам вещественных чисел с точки зрения компьютера. Во многих языках программирования поддерживаются операции над массивами, благодаря чему удобно реализовывать алгоритмы, связанные с выполнением операций над векторами.

\begin{align}

\text{Вектор столбец - a} = \begin{bmatrix}1 \\ 2 \\ 3\end{bmatrix} \in  \mathbb R ^ 3
\\
\text{Вектор строка - a} = \begin{bmatrix} 1 & 2 & 3\end{bmatrix} \in  \mathbb R ^ 3
\end{align}

### Системы линейных уравнений

Системы линейных уравнений играют центральную роль в линейной алгебре. Многие задачи можно сформулировать в виде систем линейных уравнений, а линейная алгебра предоставляет инструментарий для их решения.

Компания производит линейку продуктов $N_1, ..., N_n$, для которых требуются ресурсы $R_1, ..., R_m$. На производство единицы продукта $N_j$ требуется $ a_{ij} $ единиц ресурса $R_i$, где $i = 1, ..., m, а_j = 1, ..., n$.

Цель — подобрать оптимальный производственный план, то есть спланировать, сколько единиц $x_j$ продукта $N_j$ должно быть произведено, если доступно всего $b_i$ единиц ресурса $R_i$ и, в идеале, неизрасходованных ресурсов не остается.

Если мы произведем $x_1, ..., x_n$ единиц соответствующего продукта, то нам понадобится всего

$$ a_{ij}+\cdots+a_{in}x_n $$

единиц ресурса $R_i$. Следовательно, оптимальный производственный план $(x_1, ..., x_n) \in R_n$ должен удовлетворять следующей системе уравнений:



   \begin{equation} \tag{3}
   \begin{bmatrix}
      a_{11}x_n+\cdots+a_{1n}x_n=b_1 \\
      \vdots \\
      a_{m1}x_n+\cdots+a_{mn}x_n=b_m
   \end{bmatrix}
   \end{equation} 


где $ a_{ij} \in \mathbb R $ и $b_m \in \mathbb R ^ n $.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 40, 1000) # start, finish, n points
```

Расстояние, пройденное первым курьером: $d = 2.5t$

```python
d_r = 2.5 * t 
```

Расстояние, пройденное вторым курьером: $d = 3(t-5)$

```python
d_s = 3 * (t-5)
```

```python
fig, ax = plt.subplots()
plt.title('Встреча курьеров')
plt.xlabel('Время (минуты)')
plt.ylabel('Расстояние (км)')
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
ax.plot(t, d_r, c='green', label = 'Первый курьер')
ax.plot(t, d_s, c='brown', label = 'Второй курьер')
plt.legend()
plt.axvline(x=30, color='purple', linestyle='--')
_ = plt.axhline(y=75, color='purple', linestyle='--')
```

### Решение линейных систем

Пример, два уравнения в системе:
$$ y = 3x $$
$$ -5x + 2y = 2 $$

Второе уравнение можно перегруппировать, чтобы выделить $y$:
$$ 2y = 2 + 5x $$
$$ y = \frac{2 + 5x}{2} = 1 + \frac{5x}{2} $$


```python
x = np.linspace(-10, 10, 1000) # старт, финиш, n точек

y1 = 3 * x

y2 = 1 + (5*x)/2
```

```python
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
ax.set_xlim([0, 3])
ax.set_ylim([0, 8])
ax.plot(x, y1, c='green')
ax.plot(x, y2, c='brown')
plt.axvline(x=2, color='purple', linestyle='--')
_ = plt.axhline(y=6, color='purple', linestyle='--')
```

Пример, двумя уравнениями в системе являются:
$$ 2x - 3y = 15 $$
$$ 4x + 10y = 14 $$

Оба уравнения можно перестроить, чтобы выделить $y$. Начнем с первого уравнения:
$$ -3y = 15 - 2x $$
$$ y = \frac{15 - 2x}{-3} = -5 + \frac{2x}{3} $$

Тогда для второго уравнения:
$$ 4x + 10y = 14 $$
$$ 2x + 5y = 7 $$
$$ 5y = 7 - 2x $$
$$ y = \frac{7 - 2x}{5} $$

```python
y1 = -5 + (2*x)/3

y2 = (7-2*x)/5
```

```python
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')

# Оси x и y:
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')

ax.set_xlim([-2, 10])
ax.set_ylim([-6, 4])
ax.plot(x, y1, c='green')
ax.plot(x, y2, c='brown')
plt.axvline(x=6, color='purple', linestyle='--')
_ = plt.axhline(y=-1, color='purple', linestyle='--')
```

Для систематического подхода к решению систем линейных уравнений введем удобную компактную нотацию. Соберем коэффициенты $a_{ij}$ в векторы, а векторы — в матрицы. Иными словами, запишем систему из (3) в следующей форме:


\begin{equation} \tag{4}

   x_1
      \begin{bmatrix}
         a_{11}\\
         \vdots \\
         a_{m1}
      \end{bmatrix}
   +x_2
      \begin{bmatrix}
         a_{12}\\
         \vdots \\
         a_{m2}
      \end{bmatrix}
   + \cdots +
   x_n
      \begin{bmatrix}
         a_{1n}\\
         \vdots \\
         a_{mn}
      \end{bmatrix}
   =
      \begin{bmatrix}
         b_{1}\\
         \vdots \\
         b_{m}
      \end{bmatrix} 

\end{equation} 

\begin{equation} \tag{5}

   \Leftrightarrow
   \begin{bmatrix}{}
      a_{11} & \cdots & a_{1n}\\
      \vdots & & \vdots\\
      a_{m1} & \cdots  & a_{mn}
   \end{bmatrix}
   \begin{bmatrix}
      x_{1} \\
      \vdots\\
      x_{m}
   \end{bmatrix}
   =
   \begin{bmatrix}
      b_{1} \\
      \vdots\\
      b_{m}
   \end{bmatrix}

\end{equation} 


## Скаляры, векторы, матрицы и тензоры

Базовые определения.

<b> Разница между скаляром, вектором, матрицей и тензором. </b>

<li> Скаляр - это одно число или матрица с одной записью.

<li>  Вектор - это 1-мерный массив чисел. Другой способ представить себе вектор - это определение точек в пространстве, каждый элемент которых дает координату вдоль отдельной оси.

$$ {x} =\begin{bmatrix}
    x_1 \\\\
    x_2 \\\\
    \cdots \\\\
    x_n
\end{bmatrix}$$

<li> Матрица - это двумерный массив, в котором каждый элемент идентифицируется двумя индексами (ROW и COLUMN).

$$ {A}=
\begin{bmatrix}
    A_{1,1} & A_{1,2} & \cdots & A_{1,n} \\\\
    A_{2,1} & A_{2,2} & \cdots & A_{2,n} \\\\
    \cdots & \cdots & \cdots & \cdots \\\\
    A_{m,1} & A_{m,2} & \cdots & A_{m,n}
\end{bmatrix} $$

<li>  Тензор - это $n$ -мерный массив с $n>2$.

### Транспонирование

С помощью транспонирования возможно преобразование вектора строки в вектор столбца и наоборот.

-----

Транспонирование $A^T$ матрицы $A$ соответствует зеркальному отражению осей. Если матрица квадратная (одинаковое количество столбцов и строк).


 -----

 $$ {A}=
\begin{bmatrix}
    A_{1,1} & A_{1,2} \\\\
    A_{2,1} & A_{2,2} \\\\
    A_{3,1} & A_{3,2}
\end{bmatrix} $$



----

$${A}^{\text{T}}=
\begin{bmatrix}
    A_{1,1} & A_{2,1} & A_{3,1} \\\\
    A_{1,2} & A_{2,2} & A_{3,2}
\end{bmatrix}$$

### Сложение

Матрицы можно складывать, если они имеют одинаковую размерность:

$$ A + B = C $$

Каждая ячейка $A$ добавляется к соответствующей ячейке $B$:

$$ {A}_{ij} + {B}_{ij} = {C}_{ij}$$

$i$ - это индекс строки, а $j$ - индекс столбца.

$$ \begin{bmatrix}
    A_{1,1} & A_{1,2} \\\\
    A_{2,1} & A_{2,2} \\\\
    A_{3,1} & A_{3,2}
\end{bmatrix}+
\begin{bmatrix}
    B_{1,1} & B_{1,2} \\\\
    B_{2,1} & B_{2,2} \\\\
    B_{3,1} & B_{3,2}
\end{bmatrix}=
\begin{bmatrix}
    A_{1,1} + B_{1,1} & A_{1,2} + B_{1,2} \\\\
    A_{2,1} + B_{2,1} & A_{2,2} + B_{2,2} \\\\
    A_{3,1} + B_{3,1} & A_{3,2} + B_{3,2}
\end{bmatrix} $$

Размеры $A,B$ и $C$ одинаковы. Проверим это на примере

### Multiplying Matrices and Vectors

The standard way to multiply matrices is not to multiply each element of one with each elements of the other (this is the element-wise product) but to calculate the sum of the products between rows and columns. 

-----

The number of columns of the first matrix must be equal to the number of rows of the second matrix. Thus, if the dimensions, or the shape of the first matrix, is ($m\times n$) the second matrix need to be of shape ($n \times x$
). The resulting matrix will have the shape ($m \times x$
).


Formalization of the dot product

$$ C_{i,j} = A_{i,k}B_{k,j} = \sum_{k}A_{i,k}B_{k,j} $$

#### Пример: Рассмотрим для начала провизведение матрицы и вектора.

$$ {A} \times {b} = {C} $$

${A}=
\begin{bmatrix}
    1 & 2\\\\
    3 & 4\\\\
    5 & 6
\end{bmatrix}$ и $ {b}=\begin{bmatrix}
    2\\\\
    4
\end{bmatrix}$.

Умножение выполняется следующим образом:

$$\begin{align*}
&\begin{bmatrix}
    A_{1,1} & A_{1,2} \\\\
    A_{2,1} & A_{2,2} \\\\
    A_{3,1} & A_{3,2}
\end{bmatrix}\times
\begin{bmatrix}
    B_{1,1} \\\\
    B_{2,1}
\end{bmatrix}=\\\\
&\begin{bmatrix}
    A_{1,1}B_{1,1} + A_{1,2}B_{2,1} \\\\
    A_{2,1}B_{1,1} + A_{2,2}B_{2,1} \\\\
    A_{3,1}B_{1,1} + A_{3,2}B_{2,1}
\end{bmatrix}
\end{align*}$$

По итогу получим:

$$ \begin{align*}
&\begin{bmatrix}
    1 & 2 \\\\
    3 & 4 \\\\
    5 & 6
\end{bmatrix}\times
\begin{bmatrix}
    2 \\\\
    4
\end{bmatrix}=\\\\
&\begin{bmatrix}
    1 \times 2 + 2 \times 4 \\\\
    3 \times 2 + 4 \times 4 \\\\
    5 \times 2 + 6 \times 4
\end{bmatrix}=
\begin{bmatrix}
    10 \\\\
    22 \\\\
    34
\end{bmatrix}
\end{align*} $$

Важно проверять размерность матриц. В данном примере размер $A -$
 ($ 3 \times 2$
) и размер $b -$
 ($2 \times 1$
). Соответственно размерность $C -$
 ($3 \times 1$
).

### The Dot Product

If we have two vectors (say, ***x*** and ***y***) with the same length *n*, we can calculate the dot product between them. This is annotated several different ways, including the following: 

* $x \cdot y$
* $x^Ty$
* $\langle x,y \rangle$

Regardless which notation you use (I prefer the first), the calculation is the same; we calculate products in an element-wise fashion and then sum reductively across the products to a scalar value. That is, $x \cdot y = \sum_{i=1}^{n} x_i y_i$

The dot product is ubiquitous in deep learning: It is performed at every artificial neuron in a deep neural network, which may be made up of millions (or orders of magnitude more) of these neurons.

### Orthogonal Matrices

These are the solutions to Exercises 3 and 4 on **orthogonal matrices** from the slides.

For Exercise 3, to demonstrate the matrix $I_3$ has mutually orthogonal columns, we show that the dot product of any pair of columns is zero: 

```python
I = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])
I
```

```python
column_1 = I[:,0]
column_1
```

```python
column_2 = I[:,1]
column_2
```

```python
column_3 = I[:,2]
column_3
```

```python
np.dot(column_1, column_2)
```

```python
np.dot(column_1, column_3)
```

```python
np.dot(column_2, column_3)
```

We can use the `np.linalg.norm()` method from earlier in the notebook to demonstrate that each column of $I_3$ has unit norm: 

```python
np.linalg.norm(column_1)
```

```python
np.linalg.norm(column_2)
```

```python
np.linalg.norm(column_3)
```

Since the matrix $I_3$ has mutually orthogonal columns and each column has unit norm, the column vectors of $I_3$ are *orthonormal*. Since $I_3^T = I_3$, this means that the *rows* of $I_3$ must also be orthonormal. 

Since the columns and rows of $I_3$ are orthonormal, $I_3$ is an *orthogonal matrix*.

For Exercise 4, let's repeat the steps of Exercise 3 with matrix *K* instead of $I_3$. We could use NumPy again, but for fun I'll use PyTorch instead. (You're welcome to try it with TensorFlow if you feel so inclined.)

```python
K = torch.tensor([[2/3, 1/3, 2/3], [-2/3, 2/3, 1/3], [1/3, 2/3, -2/3]])
K
```

```python
Kcol_1 = K[:,0]
Kcol_1
```

```python
Kcol_2 = K[:,1]
Kcol_2
```

```python
Kcol_3 = K[:,2]
Kcol_3
```

```python
torch.dot(Kcol_1, Kcol_2)
```

```python
torch.dot(Kcol_1, Kcol_3)
```

```python
torch.dot(Kcol_2, Kcol_3)
```

We've now determined that the columns of $K$ are orthogonal.

```python
torch.norm(Kcol_1)
```

```python
torch.norm(Kcol_2)
```

```python
torch.norm(Kcol_3)
```

We've now determined that, in addition to being orthogonal, the columns of $K$ have unit norm, therefore they are orthonormal. 

To ensure that $K$ is an orthogonal matrix, we would need to show that not only does it have orthonormal columns but it has orthonormal rows are as well. Since $K^T \neq K$, we can't prove this quite as straightforwardly as we did with $I_3$. 

One approach would be to repeat the steps we used to determine that $K$ has orthogonal columns with all of the matrix's rows (please feel free to do so). Alternatively, we can use an orthogonal matrix-specific equation from the slides, $A^TA = I$, to demonstrate that $K$ is orthogonal in a single line of code: 

```python
torch.matmul(K.T, K)
```

Notwithstanding rounding errors that we can safely ignore, this confirms that $K^TK = I$ and therefore $K$ is an orthogonal matrix.
